```python
df = pd.read_csv("/content/2. Cars Data1.csv")
```

```python
# Finding nulls in cylinders column in the dataset and replacing them with the mean values
df.isnull().sum()
```

```python
df['Cylinders'].fillna(df['Cylinders'].mean(), inplace = True)
```

```python
df.head()
```

```python
# Types of makes in the dataset and the frequency of appearance of each make
df['Make'].value_counts()
```

```python
# Selecting cars of European or Asian Origin
df[df['Origin'].isin(['Asia', 'Europe'])]
```

```python
# Eliminating records with weight over 4000
df[~(df['Weight'] > 4000)] # Adding the ~ removes all the unwanted records.
```

```python
# Expand the values of the 'MPG_City' column by 3
df['MPG_City'] = df['MPG_City'].apply(lambda x:x+3)
df.head()
```
